```python
firRank=df.firstRank.dropna() #过滤NAN数据（存在的情况下）
firRank=[int(i.strip().replace('（首发成团位）','').replace('（淘汰）','').replace('（旁听生）','')) for i in firRank.values]
print(max(firRank),min(firRank))
```

```python
firRankcut=pd.cut(firRank,[0,7,20,55,101],labels=[u"首发成团位",u"8-20",u"21-55",u"56-101"]) #注意切分时做左边为开区间所以为0不为1
```

```python
for i,j in firRankcut.value_counts().items():
    print(i,j)
```

```python
from pyecharts import Pie,Bar,Line
attr=[]
count=[]
for i,j in firRankcut.value_counts().items():
    attr.append(i)
    count.append(j)
pie=Pie("《创造营2020》","第一次排名分布")
pie.add("第一次排名分布",attr,count,is_label_show=True)
pie.render("firstRank.html")
line=Line("《创造营2020》","第一次排名分布",width=1200,height=600)
line.add("第一次排名分布",attr,count,mark_point=['max'],is_datazoom_show=True,is_fill=True,line_opacity=0.2,area_opacity=0.4)
line.render("第一次排名分布.html")
```
